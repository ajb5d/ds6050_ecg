```python
import wfdb
import matplotlib.pyplot as plt
```

```python
rec = wfdb.rdrecord('../data/mimic/43912213')
```

```python
fig, ax = plt.subplots(12, 1, figsize=(12, 4))
for fig_i, sig in enumerate(rec.p_signal.T):
    ax[fig_i].plot(sig)
    ax[fig_i].set_ylabel(rec.sig_name[fig_i])
    ax[fig_i].set_xticks([])
    ax[fig_i].set_yticks([])

plt.savefig('figure-example-ecg.eps', format='eps')
```
