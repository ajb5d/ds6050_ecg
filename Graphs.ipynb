```python
def plot_1(df_name): 

    plt.figure(figsize=(10, 16))
    plt.subplot(3, 1, 1)
    plt.plot(df_name['epoch'], df_name['mae'], label='Training MAE')
    plt.plot(df_name['epoch'], df_name['val_mae'], label='Validation MAE')
    plt.legend(loc='best')
    plt.title('Training and Validation MAE')
    plt.xlabel('Epochs')

    plt.subplot(3, 1, 2)
    plt.plot(df_name['epoch'], df_name['loss'], label='Training Loss')
    plt.plot(df_name['epoch'], df_name['val_loss'], label='Validation Loss')
    plt.legend(loc='best')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    
    plt.subplot(3, 1, 3)
    plt.plot(df_name['epoch'], df_name['mse'], label='Training MSE')
    plt.plot(df_name['epoch'], df_name['val_mse'], label='Validation MSE')
    plt.legend(loc='best')
    plt.title('Training and Validation MSE')
    plt.xlabel('Epochs')
    
    plt.subplots_adjust(wspace=1.4)
    plt.show()
```

```python
def plot_2(df_name): 

    plt.figure(figsize=(10, 16))
    plt.subplot(3, 1, 1)
    plt.plot(df_name['epoch'], df_name['accuracy'], label='Training Accuracy')
    plt.plot(df_name['epoch'], df_name['val_accuracy'], label='Validation Accuracy')
    plt.legend(loc='best')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')

    plt.subplot(3, 1, 2)
    plt.plot(df_name['epoch'], df_name['auc'], label='Training AUC')
    plt.plot(df_name['epoch'], df_name['val_auc'], label='Validation AUC')
    plt.legend(loc='best')
    plt.title('Training and Validation AUC')
    plt.xlabel('Epochs')
    
    plt.subplot(3, 1, 3)
    plt.plot(df_name['epoch'], df_name['loss'], label='Training Loss')
    plt.plot(df_name['epoch'], df_name['val_loss'], label='Validation Loss')
    plt.legend(loc='best')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    
    plt.subplots_adjust(wspace=1.4)
    plt.show()
```
